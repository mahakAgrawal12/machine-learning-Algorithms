# Different ways of read data frame

```python
#using python dictionary
weather={
    'date':['05-01-2016','06-01-2016'],
    'maxtemp':[42,45],
    'mintemp':[33,34]
}
df=ps.DataFrame(weather)
df
```

```python
#using tuple
weather=[
    ('05-01-2016',42,33),
    ('06-01-2016',45,34)
]
df=ps.DataFrame(weather,columns=['date','maxtemp','mintemp'])
df
                                
```

# fillna(), dropna() and interpolate()

```python
newdf=df.fillna(0)
newdf
```

```python
fdf=df.fillna(method='ffill') #forward fill
fdf
```

```python
df
```

```python
bdf=df.fillna(method='bfill') #backward fill
bdf
```

```python
ldf=df.fillna(method='ffill',limit=1)
ldf
```

```python
df
```

```python
idf=df.interpolate() # take average of last and future for continous varible
idf
```

```python
df
```

```python
sdf=df.set_index('date')
fdf=sdf.dropna(how='all')
fdf
```

```python
ddf=df.dropna(thresh=3) # will keep atleast 3 non-NA values rows
ddf
```
